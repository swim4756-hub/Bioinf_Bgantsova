Для проверки наличия ортолога у человека по известной аминокислотной последовательности логично использовать blastp.

База данных RefSeq proteins предпочтительнее nr, так как она менее зашумлена и содержит меньше дубликатов. Поиск следует ограничить Homo sapiens.

tblastn пригодился бы, если бы запросом был белок, а базой — нуклеотидные последовательности (например, при неточной или неполной аннотации белка).

blastp напрямую сравнивает белок с аннотированными человеческими белками, что даёт более надёжную и простую интерпретацию гомологии.

image.png

Наиболее близким человеческим ортологом к исходному белку является родопсин (ген RHO).

Параметры выравнивания: 92% покрытие запроса, идентичность 81.82%, E-value = 2e-24. Таким образом, анализируемая последовательность представляет собой фрагмент родопсина.

У человека мутации в гене RHO ассоциированы с патологиями сетчатки, в первую очередь с пигментным ретинитом, а также с некоторыми формами врождённой стационарной ночной слепоты.

3.2 - Поиск люциферазы в транскриптоме светлячка

1. **`makeblastdb -in sequence.fasta -dbtype prot`**  
   Создаёт индексную базу данных из белковых последовательностей для быстрого поиска BLAST.

2. **`blastp -query luc.fasta.txt -db sequence.fasta -outfmt 6 > blast_hits.txt`**  
   Ищет белки в базе, похожие на запрос, результат в виде таблицы.

3. **`head -10 blast_hits.txt | cut -f2 > top10_ids.txt`**  
   Берёт 10 лучших хитов, вырезает их идентификаторы (2-й столбец), сохраняет в файл.

4. **`seqtk subseq sequence.fasta top10_ids.txt > final_candidates.fasta`**  
   Извлекает из исходного FASTA последовательности по списку идентификаторов.

5. **`grep -c ">" final_candidates.fasta`**  
   Проверяет, что извлечено 10 записей (считает заголовки).
